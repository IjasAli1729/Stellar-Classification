```python
from numpy import loadtxt
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
dataset = pd.read_csv(r"star_classification.csv")
```

```python
dataset 
```

```python
dataset.isna().sum()
```

```python
y = dataset['class']
```

```python
X = dataset.drop(['class'], axis= 1)
```

```python
X.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
y
```

```python
label_encoder = LabelEncoder()
y= label_encoder.fit_transform(y)
```

```python
X_train, X_test, y_train , y_test = train_test_split(X,y, test_size = 0.25, random_state = 42)
```

```python
input_layer = Input(shape=(17,))
hidden_layer1 = Dense(128, activation = 'relu')(input_layer)
hidden_layer2 = Dense(64, activation = 'relu')(hidden_layer1)
hidden_layer3 = Dense(32, activation = 'relu')(hidden_layer2)
hidden_layer4 = Dense(16, activation = 'relu')(hidden_layer3)
output_layer = Dense(3, activation = 'softmax')(hidden_layer4)
```

```python
model = Model(inputs = input_layer, outputs = output_layer)
model.compile(loss = 'sparse_categorical_crossentropy',
              optimizer = 'adam',
              metrics = ['accuracy']
             )
```

```python
model.fit(X_train,y_train, epochs = 100, batch_size =  150)
```

```python
_, accuracy = model.evaluate(X_test,y_test)
print("Accuracy of the dataset : %.2f" % (accuracy%100))
```
